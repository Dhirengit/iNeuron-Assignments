```python
7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to
its next() method: 2, 3, 5, 7, 11, ...
```

```python
def prime_generator(end):
    for n in range(2, end):    
        for x in range(2, n):   
            if n % x == 0:      
                break
        else:                  
            yield n             


g = prime_generator(100)  
print(list(g)) 
```
